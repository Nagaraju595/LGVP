```python
# we will consider only these columns 
data= data[['iyear', 'imonth', 'iday','country_txt','region_txt',"success","city","attacktype1_txt","targtype1_txt",
       "gname","weaptype1_txt","nkill","nwound"]]
data
```

```python
# filling the NaN values
data['nkill']=data['nkill'].fillna(0)
data['nwound']=data['nwound'].fillna(0)
```

```python
# creating a new column casualty
data['casualty']=data['nkill']+data['nwound']
```

# Visualisation of Data

```python
#Number of attacks each year

Number_of_Attacks=data["iyear"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"Year","iyear":"Attacks"}).set_index("Year")
Number_of_Attacks.head()
```

```python
plt.figure(figsize=(25,15))
plt.title('Attacks',fontweight='bold',size=20)
sns.countplot(x=data.iyear,palette="YlOrRd")
plt.xlabel("Year",fontsize=20,fontweight="bold")
plt.ylabel("Number of Attacks",fontsize=20,fontweight="bold")
plt.xticks(rotation=90,fontweight="bold",size=15)
plt.yticks(fontweight="bold",size=15)
plt.show()
```

```python
data["region_txt"].value_counts()
```

```python
plt.figure(figsize=(25,15))
plt.title('Total number of attacks Region-wise',fontweight='bold',size=20)
sns.countplot(x=data.region_txt)
plt.xlabel("Region-wise",fontsize=20,fontweight="bold")
plt.ylabel("Number of Attacks",fontsize=20,fontweight="bold")
plt.xticks(rotation=90,fontweight="bold",size=15)
plt.yticks(fontweight="bold",size=15)
plt.show()
```

✤Middle East and North Africa followed by South Asia are the hot zones for terrorism.



✤Central Asia has least number of Terrorist Attacks

```python
region=pd.crosstab(data.iyear,data.region_txt).plot(kind='area',figsize=(20,10))
plt.title('Terrorism by region in each year')
plt.ylabel('Attacks')
plt.show()
```

```python
data["country_txt"].value_counts()[:20]
```

```python
plt.figure(figsize=(25,15))
plt.title('Total number of attacks Country-wise',fontweight='bold',size=20)
sns.barplot(x=data['country_txt'].value_counts()[:20].index,y=data['country_txt'].value_counts()[:20].values)
plt.xlabel("Country-wise",fontsize=20,fontweight="bold")
plt.ylabel("Number of Attacks",fontsize=20,fontweight="bold")
plt.xticks(rotation=90,fontweight="bold",size=15)
plt.yticks(fontweight="bold",size=15)
plt.show()
```

✤Iraq has most number of terrorist attacks

```python
# Top 5 cities with highest terrorism attacks.

data['city'].value_counts().nlargest(6)
```
